```python
def missing_values(x):
    nulls=x.isnull().sum(axis=0).reset_index()
    nulls.columns=['columns','missingvalues']
    nulls=nulls[nulls['missingvalues']>0]
    nulls = nulls.sort_values(by='missingvalues')
    return nulls    
```

```python
def check_nulls(df):
    nulls = df.isnull().sum(axis=0).reset_index()
    nulls.columns = ['column', 'missing']
    nulls = nulls[nulls['missing']>0]
    nulls = nulls.sort_values(by='missing')
    return nulls    


check_nulls(traindata)
```
